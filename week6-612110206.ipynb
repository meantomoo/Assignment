```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv')
df = df.set_index('hn')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum(axis=0)
```

```python
df_clean_column = df.drop(['appearance','Wardsize','RHlvsi','depth','nodeyiel','vgmargin','pelvicme','adnmet'],axis = 1)
df_clean_column.isnull().sum(axis=0)
```

```python
df_clean = df_clean_column.dropna(axis = 0)
```

```python
df_clean.shape
```

```python
pel_class = np.unique(df_clean['pmmet'])
pel_class
```

```python
pel_value = [sum(df_clean['pmmet']==pel_class[0]),
             sum(df_clean['pmmet']==pel_class[1])]
pel_value
```

```python
y = df_clean['pmmet'].values
X = df_clean.drop(['pmmet'],axis = 1).values
```

```python
np.shape(y)
```

```python
np.shape(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

# train = 80% test = 20%
1. random_state=0
> fix random ให้มีการ random ช่วงที่ ...

```python
np.shape(X_train)
```

```python
np.shape(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

fit = การเรียน
1. feature_range=(0, 1
> เรียนรู้จากช่วง 0-1

```python
np.max(X_test_norm,0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

params = {'C': [1, 2, 4, 8, 16]}
clf = GridSearchCV(LogisticRegression(random_state=1, solver='liblinear'),params, cv=10)
clf.fit(X_train_norm, y_train)
print("Best params : " + str(clf.best_params_))
print("10CV accuracy : "+str(clf.best_score_*100))

# import sklearn as sk
# from sklearn.ensemble import RandomForestClassifier

# RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
# RF.fit(X, y)
# RF.predict(X.iloc[460:,:])
# round(RF.score(X,y), 4)
```

GridSearchCV เอาไว้จูน model

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
y_predict = clf.predict(X_test_norm)
print("Test accuracy : "+str(sum(y_test == y_predict)/len(y_test)*100))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
target_names = ['negative', 'positive']
C = confusion_matrix(y_test,y_predict) 
C = C / C.astype(np.float).sum(axis=1)*100
sns.heatmap(C, annot=True, fmt="f",cmap="GnBu",xticklabels=target_names, yticklabels=target_names)
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()
```

```python
print(classification_report(y_test, y_predict, target_names=target_names))
```
